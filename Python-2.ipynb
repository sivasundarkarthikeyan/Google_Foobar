```python
def solution(i):
    
    
    def is_prime(element, primes):
        #method is_prime is to check if the given element is a prime
        #if the element is a prime, this method adds the element to the set of primes
    
        for prime in primes:
            if not (element == prime or element % prime):
                return False
        primes.add(element)
        return True

    def get_primes(N):
        #method get_primes is to fetch first 'N' prime elements
        #in turn calls is_prime method with new element to verify if it is a prime 
        #continues to verify next element
        #return a sorted list of 'N' primes
        
        primes = set([2])
        current_element = 2
        count = 0
        while True:
            if is_prime(current_element, primes):
                count += 1
                if count == N:
                    return sorted(list(primes))
            current_element += 1
    
    #generating the first 2400 primes and concatenate them to form a string
    #concatenation will result in a string of length 10574
    primes = get_primes(2400)
    str_primes = ''
    for prime in primes:
        str_primes += str(prime)
    
    #return_data carries the new ID number
    return_data = str_primes[i:i+5]
    
    return return_data
```

```python
def solution(x, y):
    #This method returns the ID string of the bunny worker at the location x, y
    
    def natural_sum(n):
        #This method returns sum of first n natural numbers for a given n
        
        return_data = ((n*(n+1))/2)
        return return_data
    
    def match_id_with_location(m, n):
        #This method returns the bunny worker ID using given m and n
        
        row_sum = natural_sum(n)
        col_sum = natural_sum(m) - natural_sum(n-1)
        return_data = str(row_sum + col_sum)
        return return_data
    
    m = x + y - 2
    n = x
    return_data = match_id_with_location(m, n)
    return return_data
```

Gearing Up for Destruction
==========================

As Commander Lambda's personal assistant, you've been assigned the task of configuring the LAMBCHOP doomsday 
device's axial orientation gears. It should be pretty simple -- just add gears to create the appropriate rotation 
ratio. But the problem is, due to the layout of the LAMBCHOP and the complicated system of beams and pipes supporting it, the pegs that will support the gears are fixed in place.

The LAMBCHOP's engineers have given you lists identifying the placement of groups of pegs along various support beams. 
You need to place a gear on each peg (otherwise the gears will collide with unoccupied pegs). The engineers have plenty of gears in all different sizes stocked up, so you can choose gears of any size, from a radius of 1 on up. Your goal is to build a system where the last gear rotates at twice the rate (in revolutions per minute, or rpm) of the first gear, no matter the direction. Each gear (except the last) touches and turns the gear on the next peg to the right.

Given a list of distinct positive integers named pegs representing the location of each peg along the support beam, write a function solution(pegs) which, if there is a solution, returns a list of two positive integers a and b representing the numerator and denominator of the first gear's radius in its simplest form in order to achieve the goal above, such that radius = a/b. The ratio a/b should be greater than or equal to 1. Not all support configurations will necessarily be capable of creating the proper rotation ratio, so if the task is impossible, the function solution(pegs) should return the list [-1, -1].

For example, if the pegs are placed at [4, 30, 50], then the first gear could have a radius of 12, the second gear could have a radius of 14, and the last one a radius of 6. Thus, the last gear would rotate twice as fast as the first one. In this case, pegs would be [4, 30, 50] and solution(pegs) should return [12, 1].

The list pegs will be given sorted in ascending order and will contain at least 2 and no more than 20 distinct positive integers, all between 1 and 10000 inclusive.

```python
from fractions import Fraction

def solution(pegs):
    #This method returns the numerator and denominator of first's gears radius in a list when solution exists
    
    def compute_last_gear_radius(distances):
        #This method computes the last gear's radius
        #This uses the substition method for solving the linear equation
        
        last_gear_radius = 0.0
        for i in range(len(distances)):
            last_gear_radius += ((-1)**i * distances[i])

        if len(distances)%2:
            last_gear_radius = last_gear_radius/3
        
        return_data = last_gear_radius
        return return_data
    
    def compute_all_gear_radii(last_gear_radius, distances):
        #This method uses the distances between pegs and last gear's radius  
        #to compute the radius all the remaining gears
        
        all_radii = [0]*(len(distances)+1)
        all_radii[-1] = last_gear_radius
        
        for i in range(len(distances), 0, -1):
            all_radii[i-1] = distances[i-1] - all_radii[i]
        return all_radii
    
    def format_result(x):
        #This method formats the first gear's radius value as fraction
        
        fraction_x = Fraction(x).limit_denominator()
        a, b = fraction_x.numerator, fraction_x.denominator
        return_data = (a, b)
        return return_data
    
    return_data = [-1, -1]
    distances = [pegs[i+1]-pegs[i] for i in range(len(pegs)-1)]
    last_gear_radius = compute_last_gear_radius(distances)
    
    if last_gear_radius >= 1:
        all_gears_radii = compute_all_gear_radii(last_gear_radius, distances)
        
        if all(radius >= 1 for radius in all_gears_radii):
            a, b = format_result(all_gears_radii[0])
            
            if (a/b):
                return_data = [a, b]

    return return_data
```

Bomb, Baby!

```python
def solution(x, y):
    #This method returns a positive integer numerator and denominator of first's gears radius in a list when solution exists
    required_generations = 0
    next_loop = True
    m = int(x)
    f = int(y)

    while next_loop:
        current_cycle = ((max(m,f)-min(m,f))//min(m,f))
        next_loop = (max(m,f)-min(m,f))%min(m,f) > 0
        current_cycle = current_cycle + next_loop
        required_generations += current_cycle
        m, f = (max(m,f) - (current_cycle * min(m,f)), min(m,f))
            
    if m==f==1:
        return_data = str(required_generations + max(m,f) -1)
    else:
        return_data = 'impossible'
        
    return return_data
```

Doomsday Fuel
=============

Making fuel for the LAMBCHOP's reactor core is a tricky process because of the exotic matter involved. It starts as raw ore, then during processing, begins randomly changing between forms, eventually reaching a stable form. There may be multiple stable forms that a sample could ultimately reach, not all of which are useful as fuel. 

Commander Lambda has tasked you to help the scientists increase fuel creation efficiency by predicting the end state of a given ore sample. You have carefully studied the different structures that the ore can take and which transitions it undergoes. It appears that, while random, the probability of each structure transforming is fixed. That is, each time the ore is in 1 state, it has the same probabilities of entering the next state (which might be the same state).  You have recorded the observed transitions in a matrix. The others in the lab have hypothesized more exotic forms that the ore can become, but you haven't seen all of them.

Write a function solution(m) that takes an array of array of nonnegative ints representing how many times that state has gone to the next state and return an array of ints for each terminal state giving the exact probabilities of each terminal state, represented as the numerator for each state, then the denominator for all of them at the end and in simplest form. The matrix is at most 10 by 10. It is guaranteed that no matter which state the ore is in, there is a path from that state to a terminal state. That is, the processing will always eventually end in a stable state. The ore starts in state 0. The denominator will fit within a signed 32-bit integer during the calculation, as long as the fraction is simplified regularly. 

For example, consider the matrix m:
[
  [0,1,0,0,0,1],  # s0, the initial state, goes to s1 and s5 with equal probability
  [4,0,0,3,2,0],  # s1 can become s0, s3, or s4, but with different probabilities

So, we can consider different paths to terminal states, such as:

So, putting that together, and making a common denominator, gives an answer in the form of
[s2.numerator, s3.numerator, s4.numerator, s5.numerator, denominator] which is
[0, 3, 2, 9, 14].

```python
from fractions import Fraction

def solution(m):
    #This method returns an list of non-negative integers which represents the probabilities of reaching the 
    #final states in the input from the initial state s0
    
    def fraction(numerator, denominator=1):
        if numerator == 0:
            return_data = 0  
        else:
            return_data = Fraction(numerator, denominator)
        return return_data


    def matrix_subtract(a, b):
        
        n = range(len(a))
        return_data = [[a[i][j] - b[i][j] for j in n] for i in n]
        
        return return_data


    def matrix_identity(m):
        
        n = range(len(m))
        return_data = [[1 if i == j else 0 for j in n] for i in n]

        return return_data


    def matrix_multiply(a, b):
        
        return_data =  [[sum(i * j for i, j in zip(r, c)) for c in zip(*b)] for r in a]
        return return_data

    def matrix_invert(a):
        b = matrix_identity(a)
        
        for d in range(len(a)):
            t1 = fraction(1, a[d][d])
            
            for j in range(len(a)):
                a[d][j] *= t1
                b[d][j] *= t1
                
            for i in range(len(a))[0:d] + range(len(a))[d + 1 :]:
                t0 = a[i][d]
                
                for j in range(len(a)):
                    a[i][j] = a[i][j] - t0 * a[d][j]
                    b[i][j] = b[i][j] - t0 * b[d][j]
                    
        return_data =  b
        return return_data

    def normalize(probabilities):
    
        def compute_hcf(x, y):

            if x > y:
                smaller = y
            else:
                smaller = x
            for i in range(1, smaller+1):
                if((x % i == 0) and (y % i == 0)):
                    hcf = i 
            
            return_data = hcf
            return return_data

        def compute_lcm(denominator):
            first = 0
            z = list(set(denominator))
            while (len(z)>1):
                x, y = z.pop(first), z.pop(first)
                numerator = x * y
                denominator = compute_hcf(x, y)
                lcm_x_y = numerator/denominator
                z.insert(first, lcm_x_y)
            
            return_data = z[first]
            return return_data


        numerator = []
        denominator = []
        
        for probability in probabilities:
            numerator.append(probability.numerator)
            denominator.append(probability.denominator)
        
        lcm = compute_lcm(denominator[:])
        multiplier = [lcm/x for x in denominator]
        final_probability = [a*b for a,b in zip(multiplier, numerator)]
        final_probability.append(lcm)
        return_data = final_probability
        
        return return_data
    
    # Problem can be translated to an Absorbing Markov Chain 
    
    final_states = [False if sum(row) else True for row in m]

    if final_states.count(True) == 1:
        return_data =  [1, 1]
        return return_data

    p = [
        [
            1
            if final_states[state] and state == next_state
            else fraction(prob, sum(m[state]))
            for next_state, prob in enumerate(probs)
        ]
        for state, probs in enumerate(m)
    ]

    q = [
        [
            p[i][j] for j, is_final_states in enumerate(final_states) if not is_final_states]
            for i, is_final_states in enumerate(final_states)
            if not is_final_states
    ]

    r = [
        [
            p[i][j] for j, is_final_states in enumerate(final_states) if is_final_states]
            for i, is_final_states in enumerate(final_states)
            if not is_final_states
    ]

    b = matrix_multiply(matrix_invert(matrix_subtract(matrix_identity(q), q)), r)[0]
    
    return_data = normalize(b)
    
    return return_data
```
